# 1. store and time dimensional financial analytics

# 1. 1 Find top 10 district which has the most sale (in price) in 2019. 

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
select_stmt = "SELECT s.district ,SUM(t.total_price) " \
                  "FROM ecomdb_star_schema.fact_table t " \
                  "JOIN ecomdb_star_schema.store_dim s on s.store_key=t.store_key "\
                "JOIN ecomdb_star_schema.time_dim tim on tim.time_key=t.time_key "\
                "WHERE tim.year=2019 "\
                "GROUP BY CUBE(s.district) " \
                "ORDER BY SUM(t.total_price) desc "
cur.execute(select_stmt)
records = cur.fetchall()
records
```

```python
df_q1 = pd.DataFrame(records, columns=['district', 'price'])
df_q1=df_q1.dropna()
#from per item we will pick one
df_q1=df_q1.head(10)
df_q1
```

# 1.2 Year wise store sale information from 2018

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
select_stmt = "SELECT tim.year, s.store_key , SUM(t.total_price) " \
                  "FROM ecomdb_star_schema.fact_table t " \
                  "JOIN ecomdb_star_schema.store_dim s on s.store_key=t.store_key "\
                "JOIN ecomdb_star_schema.time_dim tim on tim.time_key=t.time_key "\
  "WHERE tim.year>=2018 "\
                "GROUP BY CUBE( s.store_key, tim.year) " \
                "ORDER BY s.store_key, tim.year "
cur.execute(select_stmt)
records = cur.fetchall()
records
```

```python
df_q1 = pd.DataFrame(records, columns=['year', 'store_id', 'price'])
df_q1=df_q1.dropna()
#per one store top 3 are taking
#per store id theke 3 ta 3 ta kore nibo
#df_q1=df_q1.groupby('store_id').head(3)
df_q1
```

# 2. customer and time dimensional financial analytics,

# 2.1 Find top 3 customers of 2018

```python
con = PostgresConnection().getConnection()
cur = con.cursor()

select_stmt = "SELECT cus.customer_key, cus.name, Sum(t.total_price) as totalSpent " \
                  "FROM ecomdb_star_schema.fact_table t " \
                  "JOIN ecomdb_star_schema.customer_dim cus on cus.customer_key=t.customer_key "\
                "JOIN ecomdb_star_schema.time_dim tim on tim.time_key=t.time_key "\
  "WHERE tim.year=2018 "\
                "GROUP BY CUBE( cus.customer_key, cus.name) " \
            "ORDER BY totalSpent desc "


cur.execute(select_stmt)
records = cur.fetchall()
records
```

```python
df_q1 = pd.DataFrame(records, columns=['customer_key', 'name', 'totalSpent'])
df_q1=df_q1.dropna()
#per one store top 3 are taking
#per store id theke 3 ta 3 ta kore nibo
df_q1=df_q1.head(3)
df_q1
```

# 2.2 For each customer year wise sale

```python
con = PostgresConnection().getConnection()
cur = con.cursor()

select_stmt = "SELECT cus.customer_key, cus.name, tim.year, Sum(t.total_price) as totalSpent " \
                  "FROM ecomdb_star_schema.fact_table t " \
                  "JOIN ecomdb_star_schema.customer_dim cus on cus.customer_key=t.customer_key "\
                "JOIN ecomdb_star_schema.time_dim tim on tim.time_key=t.time_key "\
                "GROUP BY CUBE( cus.customer_key, cus.name, tim.year) " \
            "ORDER BY (cus.customer_key, tim.year) "


cur.execute(select_stmt)
records = cur.fetchall()
records
```

# 3.item and time dimensional financial analytics

# 3.1 Fine top 3 products with best price sale in 2019

```python
con = PostgresConnection().getConnection()
cur = con.cursor()

select_stmt = "SELECT it.item_key, it.item_name, Sum(t.total_price) as pricee " \
                  "FROM ecomdb_star_schema.fact_table t " \
                  "JOIN ecomdb_star_schema.item_dim it on it.item_key=t.item_key "\
                "JOIN ecomdb_star_schema.time_dim tim on tim.time_key=t.time_key "\
                "GROUP BY CUBE( it.item_key, it.item_name) " \
            "ORDER BY (pricee) desc"


cur.execute(select_stmt)
records = cur.fetchall()
records
```

```python
df_q1 = pd.DataFrame(records, columns=['item_key', 'name','price'])
df_q1=df_q1.dropna()
df_q1=df_q1.head(3)
df_q1
```

# 3.2 year wise item sale. For each product find top 3 years of sale. 

```python
con = PostgresConnection().getConnection()
cur = con.cursor()

select_stmt = "SELECT it.item_key, it.item_name, tim.year, Sum(t.total_price) as pricee " \
                  "FROM ecomdb_star_schema.fact_table t " \
                  "JOIN ecomdb_star_schema.item_dim it on it.item_key=t.item_key "\
                "JOIN ecomdb_star_schema.time_dim tim on tim.time_key=t.time_key "\
                "GROUP BY CUBE( it.item_key, it.item_name, tim.year) " \
            "ORDER BY (it.item_key, SUM(t.total_price))  desc"


cur.execute(select_stmt)
records = cur.fetchall()
records
```

# 4. store and time dimensional inventory analytics,

# 4.1 Find top store of 2018 with most quantity sale. 

```python
con = PostgresConnection().getConnection()
cur = con.cursor()

select_stmt = "SELECT it.store_key,  Sum(t.quantity)" \
                  "FROM ecomdb_star_schema.fact_table t " \
                  "JOIN ecomdb_star_schema.store_dim it on it.store_key=t.store_key "\
                "JOIN ecomdb_star_schema.time_dim tim on tim.time_key=t.time_key "\
"WHERE tim.year=2018 " \
                "GROUP BY CUBE( it.store_key) " \
            "ORDER BY SUM(t.quantity)  desc"


cur.execute(select_stmt)
records = cur.fetchall()
records
```

```python
df_q1 = pd.DataFrame(records, columns=['store_key', 'quantity'])
df_q1=df_q1.dropna()
df_q1=df_q1.head(1)
df_q1
```

# 4.2 Top division of 2020 in quantity

```python
con = PostgresConnection().getConnection()
cur = con.cursor()

select_stmt = "SELECT it.division,  Sum(t.quantity)" \
                  "FROM ecomdb_star_schema.fact_table t " \
                  "JOIN ecomdb_star_schema.store_dim it on it.store_key=t.store_key "\
                "JOIN ecomdb_star_schema.time_dim tim on tim.time_key=t.time_key "\
"WHERE tim.year=2020 " \
                "GROUP BY CUBE( it.division) " \
            "ORDER BY SUM(t.quantity)  desc"


cur.execute(select_stmt)
records = cur.fetchall()
records
```

```python
df_q1 = pd.DataFrame(records, columns=['division', 'quantity'])
df_q1=df_q1.dropna()
df_q1=df_q1.head(1)
df_q1
```
