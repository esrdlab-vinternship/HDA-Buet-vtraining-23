# 1. store and time dimensional financial analytics,
***Q1: How does the average unit price of products vary across different time periods(month) at each store in every division?***


```python
con = PostgresConnection().getConnection()
curr = con.cursor()
select_stmt = '''
  SELECT t.month, s.store_key, sd.division, AVG(f.unit_price) AS avg_unit_price
FROM star_schema.fact_table f
JOIN star_schema.store_dim s ON f.store_key = s.store_key
JOIN star_schema.time_dim t ON f.time_key = t.time_key
JOIN star_schema.store_dim sd ON s.store_key = sd.store_key
GROUP BY t.month, s.store_key, sd.division

'''
curr.execute(select_stmt)
records = curr.fetchall()

records
```

```python
data = pd.DataFrame(records, columns=['month','store_key','division','avg_unit_price'])
data
```

Q2. Which store has the highest total sales  district wise quarterly?

```python
select_stmt = '''
SELECT  sd.district,sd.store_key ,t.quarter,SUM(f.total_price) AS total_revenue
FROM star_schema.fact_table f
JOIN star_schema.store_dim sd ON f.store_key = sd.store_key
JOIN star_schema.time_dim t ON f.time_key = t.time_key
GROUP BY t.quarter, sd.district,sd.store_key
ORDER BY sd.district

'''
curr.execute(select_stmt)
records = curr.fetchall()
records
```

```python
data = pd.DataFrame(records, columns=['district','store_key','quarter','total_revenue'])
data
```

# 2. customer and time dimensional financial analytics
Q1.At which quarter customers spent the most?

```python
con = PostgresConnection().getConnection()
curr = con.cursor()
select_stmt = '''
SELECT cd.customer_key, cd.name,t.quarter, SUM(f.total_price) AS total_spent_amount
FROM star_schema.fact_table f
JOIN star_schema.customer_dim cd ON f.customer_key = cd.customer_key
JOIN star_schema.time_dim t ON f.time_key = t.time_key
GROUP BY t.quarter, cd.customer_key, cd.name
ORDER BY cd.customer_key,cd.name, total_spent_amount
'''
curr.execute(select_stmt)
records = curr.fetchall()
records
```

```python
data = pd.DataFrame(records, columns=['customer_key','name','quarter','total_spent_amount'])
data
```

Q2. 10 customers who spend the most in each month

```python
select_stmt = '''
SELECT t.month, cd.customer_key, cd.name AS customer_name, SUM(f.total_price) AS total_spent_amount
FROM star_schema.fact_table f
JOIN star_schema.time_dim t ON f.time_key = t.time_key
JOIN star_schema.customer_dim cd ON f.customer_key = cd.customer_key
GROUP BY t.month, cd.customer_key, cd.name
'''
curr.execute(select_stmt)
records = curr.fetchall()
records
```

```python
data = pd.DataFrame(records, columns=['month','customer_key','customer_name','total_spent_amount'])
data
```

```python
# group the data by month
data['month'] = data['month'].astype('int64')
data['total_spent_amount'] = data['total_spent_amount'].astype('float64')
top_10_customers_per_month = []
grouped_data = data.groupby('month')

for month, group in grouped_data:
    top_10_customers = group.nlargest(10, 'total_spent_amount')
    top_10_customers_per_month.append(top_10_customers)

result = pd.concat(top_10_customers_per_month)
result


```

# 3. item and time dimensional financial analytics
Q1. Most sold 10 items for last years?

```python
con = PostgresConnection().getConnection()
curr = con.cursor()
select_stmt = '''
SELECT t.year, i.item_name, SUM(f.total_price) AS total_sales
FROM star_schema.fact_table f
JOIN star_schema.item_dim i ON f.item_key = i.item_key
JOIN star_schema.time_dim t ON f.time_key = t.time_key
GROUP BY t.year, i.item_name
ORDER BY t.year DESC, total_sales DESC;
'''
curr.execute(select_stmt)
records = curr.fetchall()
records
```

```python
data = pd.DataFrame(records, columns=['year','item_name','total_sales'])
data
```

```python
# group the data by year
data['year'] = data['year'].astype('int64')
data['total_sales'] = data['total_sales'].astype('float64')
top_10_items_per_year = []
grouped_data = data.groupby('year')

for year, group in grouped_data:
    top_10_items = group.nlargest(10, 'total_sales')
    top_10_items_per_year.append(top_10_items)

result = pd.concat(top_10_items_per_year)
result
```

Q2.What is the trend of product's unit price over time(yearly)?

```python
select_stmt = '''
SELECT t.year, i.item_name, AVG(f.unit_price) AS average_unit_price
FROM star_schema.fact_table f
JOIN star_schema.item_dim i ON f.item_key = i.item_key
JOIN star_schema.time_dim t ON f.time_key = t.time_key
GROUP BY t.year, i.item_name
ORDER BY i.item_name ,t.year, average_unit_price;
'''
curr.execute(select_stmt)
records = curr.fetchall()
records
```

```python
data = pd.DataFrame(records, columns=['year','item_name','average_unit_price'])
data
```
