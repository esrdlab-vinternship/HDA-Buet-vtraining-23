### Load ###

#### Connect to DB ####

```python
class PostgresConnection(object):
    def __init__(self):
        self.connection = psycopg2.connect(database="hddb",
                                           user = "postgres",
                                           password = "1234",
                                           host = "127.0.0.1",
                                           port = "5432")

    def getConnection(self):
        print("Connection to DB established!")
        return self.connection
```

```python
con = PostgresConnection().getConnection()
```

#### Insert Data into tables #####

```python
con = PostgresConnection().getConnection()
insert_stmt = "INSERT INTO hddb_star_schema.geography_dim(geo_key, upazila, district,division)" \
            " VALUES (%s, %s, %s,%s)"
cur = con.cursor()
psycopg2.extras.execute_batch(cur, insert_stmt, geography_dim.values)
con.commit()
cur.close()
```
