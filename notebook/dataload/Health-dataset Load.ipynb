#### Insert Data into tables #####

```python
con = PostgresConnection().getConnection()
insert_stmt = "INSERT INTO healthdb_schema.\"geography_dim\"(geo_key, upazila, district, division)" \
            " VALUES (%s, %s, %s, %s)"
cur = con.cursor()
psycopg2.extras.execute_batch(cur, insert_stmt, geography_data.values)
con.commit()
cur.close()
```

```python
import time

start_time = time.time()

con = PostgresConnection().getConnection()
insert_stmt = "INSERT INTO healthdb_schema.\"time_dim\"(time_key, date_time, year, month, day, yyyymm, yyyymmdd, quarter)VALUES (%s, %s, %s, %s, %s, %s, %s, %s);"
cur = con.cursor()
psycopg2.extras.execute_batch(cur, insert_stmt, time_data.values)
con.commit()
cur.close()

end_time = time.time()

print("Excecution time: {:.2f} seconds".format(end_time - start_time))
```
